#### Problem Description
The sinking of the Titanic is one of the most infamous shipwrecks in history.
On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.
While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.
In this assignment, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (i.e., name, age, gender, socio-economic class, etc.).


## Part A: Data Import and Data Preprocessing

#### A. Read the data file

```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('Data-Hw3.csv')
x = dataset.iloc[:, 2:-1].to_numpy() # exclude passenger ID and name from dataset
y = dataset.iloc[:,-1].to_numpy()
```

#### B. Replace missing data
Column 3: Age

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:2])
x[:, 1:2] = imputer.transform(x[:, 1:2])
```

#### Convert categorical data (gender)

```python
# Col 0: gender (male/female)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=
                       [('encoder', OneHotEncoder(), [0])
                       ], remainder='passthrough')

x = np.array(ct.fit_transform(x))
```

#### C. Divide dataset into Training set and Test set
25% of data should be used for the Test set.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Part B: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_regr = LogisticRegression()
logistic_regr.fit(x_train, y_train)
```

#### A. Print the prediction and corresponding ground truth in Test set

```python
y_pred_logistic = logistic_regr.predict(x_test)
print(np.concatenate((y_pred_logistic.reshape(len(y_pred_logistic), 1), y_test.reshape(len(y_test), 1)), 1))
```

#### B. Print the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_logistic)
print(cm)
```

#### C. Compute Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_logistic)
print(accuracy)
```

#### B. What could possibly make the top 3 models outperform the rest?
**K-SVM** is able to separate points that are **not** linearly separable. The mapping to a higher dimension leads to a nonlinear separation, compared to a linear separation in regular SVM. Also, the combination of multiple kernels can lead to an even higher accuracy. **K-NN** also uses a nonlinear divider compared as opposed to a linear divider in logistic regression. This makes it a better candidate for classification as compared to logistic regression. **Random Forest classification** also uses a nonlinear "divider" by using multiple decision trees assign a point.
